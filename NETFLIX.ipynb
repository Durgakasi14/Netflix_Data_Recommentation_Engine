```python
df=df.drop(columns=["date_added"])
```

```python
for col in["director","listed_in","type","country","release_year"]:
    df[col]=df[col].fillna('')
    df["combined_feature"]=df["director"] +' '+ df["listed_in"] +' '+ df["type"] +' '+ df["country"] +' '+ df["release_year"].astype(str)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
feature_vectors=vectorizer.fit_transform(df["combined_feature"])
print(feature_vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(feature_vectors)
print(similarity)
```

```python
print(similarity.shape)
```

```python
list_of_all_titles=df["title"].tolist()
print(list_of_all_titles)
```

```python
movie_name=input("Enter the favourite movie name:")
```

```python
import difflib
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
detail_of_the_movie=df[df.title==close_match]["combined_feature"].values[0]
print(detail_of_the_movie)
```

```python
type_of_the_movie=df[df.title==close_match]["type"].values[0]
print(type_of_the_movie)
```

```python
def recommend(title,df,similarity):
    idx=df[df["title"].str.lower()==title.lower()].index[0]
    sim_scores=list(enumerate(similarity[idx]))
    sim_scores=sorted(sim_scores,key=lambda x:x[1],reverse=True)
    sim_scores=sim_scores[1:11]
    movie_indices=[i[0] for i in sim_scores]
    return df["title"].iloc[movie_indices]
fav_movie=input("Enter your fav movie")
recommend(fav_movie,df,similarity)
```

```python
recommend("The Lies within",df,similarity)
```

```python
def recommend_by_director(title,df,similarity):
    df["title_lower"]=df["title"].str.lower()
    director_list=df.loc[df["title_lower"]==title,"director"].values
    if len(director_list)==0 or pd.isna(director_list[0]):
        print("director information is not available")
        return
    director_name=director_list[0]
    same_director_movies=df[df["director"]==director_name]["title"].tolist()
    same_director_movies=[m for m in same_director_movies if m.lower()!=title]
    if len(same_director_movies)==0:
        print(f"no other movies found by director{director_name}.")
    else:
        print("movies directed by {director_name}:\n")
        for i,movie in enumerate(same_director_movies,1):
            print(f"{i}.{movie}")
fav_movie=input("Enter your fav movie")
recommend_by_director(fav_movie,df,similarity)
```
